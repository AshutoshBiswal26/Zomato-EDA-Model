```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(3)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
#Zomato maximum transaction records are from India. After that USA and UK
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observation
1. Whenever the rating are between 4.5 to 4.9 ----> excellent
2. When ratings are between 3.5 to 3.9 ----> very good
3. When ratings are between 3.0 to 3.4 -----> good

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating',y='Rating Count', hue='Rating color',data=ratings)
```

Observation
1. Count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','green','red','orange','yellow','indigo'])
```
